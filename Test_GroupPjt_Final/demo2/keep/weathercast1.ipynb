```python
# 라이브러리 가져오기
import numpy as np
import pandas as pd
```

```python
# weather 데이터 1
weather1 = pd.read_csv('./data/2000-01~2021-11m.csv', encoding='cp949')

# 날씨 데이터 1 개요
weather1.info()
```

```python
# 2000년 1월 ~ 2019년 12월
trainSet = weather1.iloc[0:240,3:8]
# 2019년 1월 ~ 2020년 12월
testSet = weather1.iloc[228:252,3:8]
```

```python
# 데이터 정규화
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
trainSet = scaler.fit_transform(trainSet)
testSet = scaler.fit_transform(testSet)
```

```python
# build datasets
def build_dataset(time_series, seq_length):
    dataX = []
    dataY = []
    for i in range(0, len(time_series) - seq_length):
        x = time_series[i:i + seq_length, :]
        y = time_series[i + seq_length, :]
        dataX.append(x)
        dataY.append(y)
    return np.array(dataX), np.array(dataY)
```

```python
seq_length = 12
trainX, trainY = build_dataset(trainSet, seq_length)
testX, testY = build_dataset(testSet, seq_length)

print(trainX.shape)
print(trainY.shape)
print(testX.shape)
print(testY.shape)
```

```python
################ relu layer 3개로 모델 형성해봤으나 학습이 나아지지 않아서 폐기
# 데이터 정규화 안했고 데이터셋 빌드 안한것인 원인인가 해서 chap08/06을 복붙해 모델 만드려고 하였으나
# 모델 생성에서 실패중

import tensorflow as tf

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.LSTM(units=60, input_shape=(12, 5)))

tf.model.add(tf.keras.layers.Dense(units=5, activation='tanh'))
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(lr=0.1), metrics='accuracy')
```

```python
tf.model.fit(trainX, trainY, epochs=100)
```
